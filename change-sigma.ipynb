```python
import GPy
```

```python
import numpy as np
```

# A Simple Model

```python
X = np.linspace(0, 10, 11).reshape((-1, 1))
```

```python
Y = np.sin(X)
```

```python
kernel = GPy.kern.RBF(input_dim=1, lengthscale=.5)
```

```python
m = GPy.models.GPRegression(X, Y, kernel, noise_var=0.1)
```

```python
fig = m.plot()
```

# Changing the Observational Noise

```python
m1 = GPy.models.GPRegression(X, Y, kernel, noise_var=1e-4)
```

```python
x_test = np.linspace(0, 10, 101).reshape((-1, 1))
```

```python
y1 = m1.predict(x_test)[0]
```

```python
m2 = GPy.models.GPRegression(X, Y, kernel, noise_var=1e-6)
```

```python
y2 = m2.predict(x_test)[0]
```

```python
np.max(np.abs(y2-y1))
```

Changing the observational variance from $10^{-4}$ to $10^{-6}$ seems to change the mean by $8.52\cdot 10^{-5}$.
